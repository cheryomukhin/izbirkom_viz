```python
with open(r'D:\work\votes\ЮАО.txt', 'r') as f:
    text = json.loads(f.read().replace('\n', ' '))['items']
    regions = []
    uiks= []
    parties = []
    candidates = []
    candidates_names = []
    for region in text:
        #print(region['region']['region_name'], '\n\n')
        region_name = region['region']['region_name']
        for i in region['districts']['candidates'].keys():
            regions.append({'region_name':region_name, 'district':i, "data":region['districts']['candidates'][i]})
        for j in region['uiks']:
            uiks.append({'region_name':region_name,'uik':j, 'votes':region['uiks'][j]['votes']})
            parties += region['uiks'][j]['votes'].keys()
            candidates.append({'region_name':region_name,'uik':j, 'votes':region['uiks'][j]['candidates']})
            candidates_names += region['uiks'][j]['candidates'].keys()
        #for uik in region['districts']['candidates'].keys():
        #    for candidate in region['districts']['candidates'][uik]:
        #        print(uik, candidate['full_name'], candidate['votes'], '\n')
#print(regions[1])
print(uiks[1])
parties = set(parties)
data_regions = json_normalize(regions, 'data', 'region_name')
print(data_regions.head())
data_uiks = json_normalize(uiks)
#data_uiks_melt = pd.melt(data_uiks, id_vars=['region_name', 'uik'], value_vars=['votes.'+i for i in parties]).dropna()
data_uiks_melt = pd.melt(data_uiks.rename(columns=lambda x: x.split('.')[-1]), id_vars=['region_name', 'uik'], value_vars=parties).dropna()
print(parties)
print(data_uiks.head())
print(data_uiks_melt.head())
```
